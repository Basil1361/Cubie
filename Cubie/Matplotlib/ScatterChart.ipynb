# We can actually use c as param instead of color

```python
import matplotlib.pyplot as plt

# Sample data
x_data = [3, 5, 2, 7, 6]
y_data = [75, 85, 60, 90, 80]
marker_sizes = [50, 100, 30, 70, 60]
colors = ['red', 'green', 'blue', 'orange', 'purple']
marker_styles = ['o', 's', '^', 'D', 'v']
linewidths = [2, 2, 2, 2, 2]
edgecolors = ['black', 'black', 'black', 'black', 'black']

# Creating the scatter plot
for i in range(len(x_data)):
    plt.scatter(x_data[i], y_data[i], s=marker_sizes[i], c=colors[i], marker=marker_styles[i], linewidths=linewidths[i], edgecolor=edgecolors[i])

# Adding labels and title
plt.xlabel('Hours of Study')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs. Exam Scores')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

brands = ['Starbucks', 'Pokka', 'Kopiko', 'Nescafe', 'Wonda', 'FrenchRoast']
sold_items_first_month = [250, 515, 189, 250, 370, 200]
sold_items_second_month = [370, 467, 200, 378, 389, 305]


plt.scatter(brands, sold_items_first_month, c = 'blue', label = "First Month")
plt.scatter(brands, sold_items_second_month, c = 'green', label = "Second Month")
plt.xlabel('Brands')
plt.ylabel('Number of Sold Items')
plt.title('Sales of Bottled Coffee Drinks for Different Brands')
plt.legend()

plt.show()
```
